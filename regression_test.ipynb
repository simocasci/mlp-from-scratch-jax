```python
import jax
import jax.numpy as jnp
from jax import vmap
from mlp import MLP
import tensorflow.keras.datasets as datasets
import numpy as np
import matplotlib.pyplot as plt
```

```python
def get_preprocessed_boston_housing_prices():
    (x_train, y_train), (x_test,y_test) = datasets.boston_housing.load_data(path="boston_housing.npz",
                                                                            test_split=0.1,
                                                                            seed=0)
    train_mean = np.mean(x_train, axis=0)
    train_std = np.std(x_train, axis=0)
    x_train = (x_train - train_mean) / train_std
    
    test_mean = np.mean(x_test, axis=0)
    test_std = np.std(x_test, axis=0)
    x_test = (x_test - test_mean) / test_std
    
    return x_train, y_train, x_test, y_test
```

```python
x_train, y_train, x_test, y_test = get_preprocessed_boston_housing_prices()
```

```python
mlp = MLP([13, 20, 1], 'regression')
```

```python
%%time
mlp.train(x_train, y_train, epochs=10)
```

```python
print(mlp.mean_squared_error(x_train, y_train))
```

```python
print(mlp.mean_squared_error(x_test, y_test))
```

```python
plt.plot(y_test)
plt.plot(mlp.predict(mlp.params, x_test))
plt.show()
```
